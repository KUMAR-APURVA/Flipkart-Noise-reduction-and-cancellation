```python
import pandas as pd
dataset = pd.read_csv('flipkart_com-ecommerce_sample.csv',engine='python')
```

```python
product_category_tree = dataset['product_category_tree']
```

```python
product_list = set()
```

```python
for elem in product_category_tree:
#     print(elem)
    elem = elem[2:-2]
    text = elem.split(' >> ')
    text = list(' '.join(text))
#     text = text.split()
    for i in range(len(text)):
        if text[i]==',' or text[i]=='.' or text[i]=='!' or text[i]=='?' \
        or text[i]==')'or text[i]==';' or text[i]==':' or text[i]=='&' or text[i]=='(':
            text[i] = ''
    text = ''.join(text)
    text = text.lower()

    text = text.split()
    for x in text:
        product_list.add(x)
    
```
